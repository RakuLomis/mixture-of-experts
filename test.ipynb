```python
import torch
```

```python
gates = torch.rand(4, 5)
print("----------gates----------") 
print(gates)
_gates = gates
_num_experts = 5
print("----------torch.nonzero(gates)----------")
print(torch.nonzero(gates))
sorted_experts, index_sorted_experts = torch.nonzero(gates).sort(0)
print("----------sorted_experts----------")
print(sorted_experts)
print("----------index_sorted_experts----------")
print(index_sorted_experts)
_, _expert_index = sorted_experts.split(1, dim=1) # indeed, it returns a tuple
print("----------sorted_experts.split(1, dim=1)----------")
print(sorted_experts.split(1, dim=1))
print("----------_expert_index----------")
print(_expert_index)
_batch_index = torch.nonzero(gates)[index_sorted_experts[:, 1], 0]
print("----------_batch_index----------")
print(_batch_index)
_part_sizes = (gates > 0).sum(0).tolist()
print("----------_part_sizes--------")
print(_part_sizes)
gates_exp = gates[_batch_index.flatten()]
print("----------gates_exp----------")
print(gates_exp)
_nonzero_gates = torch.gather(gates_exp, 1, _expert_index)
print("----------_nonzero_gates----------")
print(_nonzero_gates)
print(_nonzero_gates.shape)
```
